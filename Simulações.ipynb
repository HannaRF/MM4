## Modelo SEIRD 


$$\frac{dS}{dT}=-\lambda \cdot S(t) \cdot I(t)$$

$$\frac{dE}{dT}=\lambda \cdot S(t) \cdot I(t) - \alpha \cdot E(t) $$

$$\frac{dI}{dT}=\alpha \cdot E(t) - (\gamma + \mu) \cdot I(t)$$

$$\frac {dR}{dT}=\gamma \cdot I(t)$$

$$\frac {dD}{dT}=\mu \cdot I(t)$$

```python
def ODE (t,y,params):
    S,E,I,R,D = y
    lam,alfa,gama,mu = params
    return [-lam*S*I,
            lam*S*I - alfa*E,
            alfa*E - gama*I - mu*I,
            gama*I,
            mu*I ]
```
